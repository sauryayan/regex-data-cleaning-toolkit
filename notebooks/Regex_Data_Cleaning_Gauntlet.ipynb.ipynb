# Regular Expressions Gauntlet

```python
# Importing the library
import re
```

### Log text

```python
text2 = """
Customer Support Log - Transcript Alpha-99. > On 2026-03-15 at 14:30:00, Mr. 
John Doe (ID: Cust_8492) contacted support regarding order #AB-4921. He mentioned 
his alternate number is (555) 123-4567, but his primary contact is listed as 
555-987-6543. Another user, Dr. Jane Smith, called at 09:15 AM on 12/04/2025. 
Her phone number is +1-800-555-0199. She was inquiring about the 'Data Analytics 
Masterclass' subscription priced at $199.99. During the afternoon shift, at 16:45, a 
vendor from the Mumbai office (contact: 91-7876543210) reported a server error: ERR_TIMEOUT_504. 
The vendor's email is vendor.support_ind@tech-hub.co.in. We also received an email from a VIP 
customer, Ms. Alice Wonderland (alice.w@wonder-land.net) on 2025.10.31 complaining about a 
late delivery for order #XY-9932. Please forward all logs to admin_logs123@company.com. 
Note: system maintenance is scheduled for April 5th, 2026 at 11:00 PM. Old transaction references 
like TXN-xyz-773 and TXN-abc-192 need to be archived immediately. Total revenue impacted by the 
downtime was approximately $4,500.00.Order, order, new_email: admin.09@google.gov.in.org Mrs. Tiwari ₹45,000
"""
```

#### Category 1: Basic Matching & Identifiers

```python
# 1. Extract all standalone numbers (digits only, no letters or symbols).
re.findall(r'\d', text2)
```

```python
# 2. Extract all words containing exactly 3 letters.
pattern_3_words = r'\b[a-zA-Z]{3}\b'
re.findall(pattern_3_words, text2)
```

```python
# 3. Find all occurrences of the word "order" (make it case-insensitive so it matches "Order" or "order")
re.findall(r'order', text2, re.I)
```

```python
# 4. Extract all transaction IDs that start with "TXN" (e.g., TXN-xyz-773).
"""
- Used flag 're.I' to make the text case insensitive
"""
pattern_txn = r'txn\-?\w*-\w*'
re.findall(pattern_txn, text2, re.I)
```

```python
# 5. Find all system error codes that start with "ERR".
re.findall(r'ERR+[._]+\w+', text2)
```

```python
# 6. Extract all Customer IDs formatted as "Cust_" followed by numbers.
re.findall(r'Cust_\d+', text2)
```

```python
# 7. Extract all Order numbers starting with "#" followed by two uppercase letters and four digits.
re.findall(r'#\s?[A-Z]+-\d{4}', text2)
```

```python
# 8. Using a capture group, extract just the numeric part of the Order numbers (ignore the # and letters).
re.findall(r'#\s?[A-Z]+-(\d{4})', text2)
```

#### Category 2: Phone numbers

```python
# 9. Extract phone numbers specifically in the format (XXX) XXX-XXXX.
re.findall(r'\(\d{3}\)\s?\d{3}-\d{4}', text2)
```

```python
# 10. Extract phone numbers specifically in the format XXX-XXX-XXXX.
re.findall(r'\d{3}-\d{3}-\d{4}', text2)
```

```python
# 11. Extract phone numbers starting with a country code (e.g., +1- or 91-).
re.findall(r'(?:\+\d|[91])-[0-9-]+', text2)
```

```python
# 12. Write a single, combined regex pattern that successfully captures all four phone number formats present in the text.
"""
Phone number pattern:
(XXX) XXX-XXXX
XXX-XXX-XXXX
+X-XXX-XXX-XXXX
XX-XXXXXXXXXX (10 Digit number wit country code without '+')
"""
re.findall(r'(?:\+\d{1,3}-)?(?:\(\d{3}\)\s?|\d{3}-)\d{3}-\d{4}|\d{2}-\d{10}', text2)
```

```python
# 13. Using a capture group, extract just the area codes (the first 3 digits)
# from the US-style phone numbers.
pattern = r'(?:\+\d{1,3}-)?\(?(\d{3})\)?[\s-]\d{3}[\s-]\d{4}'
re.findall(pattern, text2)
```

#### Category 3: Dates & Times

```python
# 14. Extract dates in the strict ISO format: YYYY-MM-DD.
re.findall(r'\d{4}-\d{2}-\d{2}', text2)
```

```python
# 15. Extract dates in the US format: MM/DD/YYYY.
re.findall(r'\d{2}/\d{2}/\d{4}', text2)
```

```python
# 16. Extract dates separated by dots: YYYY.MM.DD.
re.findall(r'\d{4}\.\d{2}\.\d{2}', text2)
```

```python
# 17. Combined Date: Write a single regex to capture YYYY-MM-DD, MM/DD/YYYY, and YYYY.MM.DD.
re.findall(r'(?:\d{2}|\d{4})[-\./]\d{2}[-\./](?:\d{4}|\d{2})', text2)
```

```python
# 18. Extract times in the 24-hour format with seconds (HH:MM:SS).
re.findall(r'\d{2}:\d{2}:\d{2}', text2)
```

```python
# 19. Extract times in the 24-hour format without seconds (HH:MM).
re.findall(r'\d{2}:\d{2}', text2)
```

```python
# 20. Extract times formatted in the 12-hour format with AM/PM (e.g., 09:15 AM).
re.findall(r'\d{2}:\d{2}\s\w{2}', text2)
```

#### Category 4: Emails, Web, & Text Parsing

```python
# 21. Use a capture group to extract just the month (MM) from the YYYY-MM-DD dates.
re.findall(r'\d{4}-(\d{2})-\d{2}', text2)
```

```python
# 22. Extract all complete email addresses.
# [^.\s] at last is used to prevent capturing any white spaces and '.' at the end of string.
re.findall(r'[a-zA-z_0-9\.]+\.?[a-zA-Z0-9]@(?:[a-zA-Z0-9-]+\.[a-zA-Z0-9\.\-]+[^\.\s])', text2)
```

```python
# 23. Extract just the domain names from the email addresses (e.g., tech-hub.co.in, company.com).
re.findall(r'@([a-zA-Z0-9-]+\.[a-zA-Z0-9\.\-]+[^\W)])', text2)
```

```python
# 24. Extract the usernames from the email addresses (the part before the @).
re.findall(r'([a-zA-Z0-9-\._]+)@', text2)
```

```python
# 25. Extract all website/domain endings (e.g., .com, .net, .co.in).
re.findall(r'@[a-zA-Z0-9-]+(\.[a-zA-Z0-9\.\-]+[^\W)])', text2)
```

```python
# 26. Extract all personal titles (Mr., Ms., Dr.) along with the person's first name.
re.findall(r'[MD](?:s|r|rs)\.', text2)
```

```python
# 27. Extract the specific text enclosed strictly inside single quotes
# (e.g., 'Data Analytics Masterclass').
re.findall(r"'[a-zA-Z0-9-\s?]+'", text2)
```

#### Category 5: Money and Data masking

```python
# 28. Extract all currency amounts including the dollar sign (e.g., $199.99, $4,500.00, ₹45,000).
re.findall(r'[\$\₹]\s?[0-9,.]+\d{2}?', text2)
```

```python
# 29. Extract just the numeric values of the currency amounts (ignore the $, but keep the commas and decimals).
re.findall(r'[\$\₹]\s?([0-9,.]+\d{2}?)', text2)
```

```python
# 30. Data Privacy (Substitution): Write the regex you would use in a re.sub() or .str.replace() 
# function to find all email addresses and replace them with the string [REDACTED_EMAIL].
email_pattern = r'[a-zA-z_0-9\.]+\.?[a-zA-Z0-9]@(?:[a-zA-Z0-9-]+\.[a-zA-Z0-9\.\-]+[^\.\s])'
redacted_email_text = re.sub(email_pattern, '[REDACTED_EMAIL]', text2)
redacted_email_text
```
